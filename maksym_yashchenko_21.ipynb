Cleaning dataset

```python
df_application_cleaned = df_application.copy()
df_application_cleaned = pd.merge(df_application_cleaned, df_industries, left_on='Industry', right_on='Industry', how='left')
df_application_cleaned = pd.DataFrame(df_application_cleaned).rename(columns={'Applied at': 'applied_at',
                                                     'Amount': 'amount',
                                                     'Age': 'age',
                                                     'Gender': 'gender',
                                                     'Industry': 'industry',
                                                     'Marital status': 'marital_status',
                                                     'External Rating': 'external_rating',
                                                     'Education level': 'education_level',
                                                     'Location': 'location',
                                                     'Score': 'score'})

df_application_cleaned['applicant_id'] = df_application_cleaned['applicant_id'].drop_duplicates().reset_index(drop=True)
df_application_cleaned['applied_at'] = df_application_cleaned['applied_at'].astype("datetime64[s]")
df_application_cleaned = df_application_cleaned[['applicant_id', 'applied_at', 'amount', 'age', 'gender', 'marital_status', 'education_level', 'industry', 'external_rating', 'location', 'score']]
df_application_cleaned['external_rating'] = df_application_cleaned['external_rating'].fillna(0).astype(float)
df_application_cleaned['education_level'] = df_application_cleaned['education_level'].fillna('Середня')
df_application_cleaned['location'] = df_application_cleaned['location'].fillna('Unknown')
df_application_cleaned['amount'] = df_application_cleaned['amount'].fillna(0).astype(float)
```

Adding rating column

```python
df_application_cleaned['rating'] = int(0)

def age_rating(row) -> int:
    '''Function to calculate rating based on age'''
    if 35 < row['age'] < 55:
        return 20
    else:
        return 0

def weekend_rating(row) -> int:
    '''Function to calculate rating based on application day'''
    if row['applied_at'].dayofweek >=5:
        return 0
    else:
        return 20

def martial_status_rating(row) -> int:
    '''Function to calculate rating based on marital status'''
    if row['marital_status'] == 'Married':
        return 20
    else:
        return 0

def location_rating(row) -> int:
    '''Function to calculate rating based on location'''
    if row['location'] == 'Київ чи область':
        return 20
    else:
        return 0

def score_rating(row) -> int:
    '''Function to calculate rating based on score'''
    return row['score']

def external_rating_rating(row) -> int:
    '''Function to calculate rating based on external rating'''
    if row['external_rating'] >= 7:
        return 20
    elif row['external_rating'] <= 2:
        return 20
    else:
       return 0

def zero_rating(row) -> int:
    '''Function to set rating to zero if amount is zero'''
    if row['amount'] == 0 or row['external_rating'] == 0:
        return 0
    else:
        return row['rating']

def bound_rating(row) -> int:
    '''Function to bound rating between 0 and 100'''
    if row['rating'] > 100:
        return 100
    elif row['rating'] < 0:
        return 0
    else:
        return row['rating']

func_list = [age_rating, weekend_rating, martial_status_rating, location_rating, score_rating, external_rating_rating]

def apply_ratings(row) -> int:
    '''Function to apply all rating functions'''
    rating = 0
    for func in func_list:
        rating += func(row)
    return rating

df_application_cleaned['rating'] = df_application_cleaned.apply(axis=1, func=apply_ratings)
df_application_cleaned['rating'] = df_application_cleaned.apply(axis=1, func=zero_rating)
df_application_cleaned['rating'] = df_application_cleaned.apply(axis=1, func=bound_rating)
```

The final dataset

```python
df_final = df_application_cleaned[(df_application_cleaned['rating'] > 0)].copy().reset_index(drop=True)
df_final.head()
```

**Analysis: Average rating per week**

```python
df_grouped = pd.DataFrame(df_final[['applied_at', 'rating']].reset_index(drop=True), columns=['applied_at', 'rating'])
df_grouped['year'] = df_grouped['applied_at'].dt.year
df_grouped['weeks'] = df_grouped['applied_at'].dt.isocalendar().week
df_grouped = df_grouped.groupby(['year', 'weeks'], as_index=False).agg({'rating': 'mean'})
```
